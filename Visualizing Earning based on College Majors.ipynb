# Analysis of Earnings based on College Majors

The American Community Survey conducted a survey and aggregated data on the job outcomes of students that graduated between 2010 and 2012. FiveThirtyEight cleaned the data, and made it available on their Github.

We will use this dataset to visualize it using Python, and gain insight on:
1. whether students in popular majors make more money
2. how many majors are predominantly male or female
3. what category of majors have the most students

## Import and Explore Dataset

We'll start by reading the related dataset, and exploring some of the data.

```python
import pandas as pd
import matplotlib as plt

# run matplotlib inline so plots can be displayed inline
%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

```python
recent_grads.head()
```

```python
recent_grads.tail()
```

```python
recent_grads.describe()
```

Since matplotlib would throw an error if columns input don't match in the number of values, we'll need to remove rows with missing values.

```python
raw_data_count = recent_grads.shape[0]
print(raw_data_count)
```

```python
for x in recent_grads.columns:
    print(x, ': ', recent_grads[x].count())
```

The `Total`, `Men`, `Women`, and `Sharewomen` columns have one row each with no values. 

```python
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print(cleaned_data_count)
```

Only one row in the dataframe had missing values. 

## Scatter Plots
We'll use scatter plots to examine the correlation between datapoints.


```python
ax = recent_grads.plot(x='Median', y='Sample_size', kind='scatter')
ax.set_title('Median vs Sample-size')
```

```python
ax = recent_grads.plot(x='Unemployment_rate', y='Sample_size', kind='scatter')
ax.set_title('Unemployment-rate vs Sample-size')
```

```python
ax = recent_grads.plot(x='Median', y='Full_time', kind='scatter')
ax.set_title('Full time vs Median')
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('Unemployment-rate vs Share of Women')
```

```python
ax = recent_grads.plot(x='Median', y='Men', kind='scatter')
ax.set_title('Median vs Men')
```

```python
ax = recent_grads.plot(x='Median', y='Women', kind='scatter')
ax.set_title('Median vs Women')
```
